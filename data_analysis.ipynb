```python
df.columns = [c.replace(' ', '_').replace('-', '_') for c in df.columns]
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
if 'Review_Rating' in df.columns:
    df['Review_Rating'] = df.groupby('Category')['Review_Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100
```

```python
import sqlite3
conn = sqlite3.connect('superstore_db')
df.to_sql('sales_table', conn, if_exists='replace', index=False)
```

```python
conn = sqlite3.connect('superstore_analytics.db')
```

```python
df.to_sql('orders', conn, if_exists='replace', index=False)
```

```python
print("Table 'orders') created successfully.")
```

```python


query_region = """
SELECT Region, SUM(Sales) as Total_Sales, SUM(Profit) as Total_Profit
FROM orders
GROUP BY Region
ORDER BY Total_Profit DESC;
"""
region_df = pd.read_sql(query_region, conn)
print("--- Region Performance ---")
display(region_df)
```

```python

query_loyalty = """
SELECT Customer_ID, Customer_Name, COUNT(Order_ID) as Frequency, SUM(Sales) as Total_Revenue
FROM orders
GROUP BY Customer_ID, Customer_Name
ORDER BY Frequency DESC
LIMIT 10;
"""

loyalty_df = pd.read_sql(query_loyalty, conn)
print(loyalty_df)
```

```python
query_segments = """
SELECT Segment,
       COUNT(Order_ID) as Total_Orders,
       SUM(Sales) as Revenue,
       AVG(Discount) as Avg_Discount
FROM orders
GROUP BY Segment
ORDER BY Revenue DESC;
"""
segment_df = pd.read_sql(query_segments, conn)
print("--- Customer Segment Analysis ---")

display(segment_df)
```
